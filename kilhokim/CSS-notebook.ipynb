```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(os.path.join("images", "ssh.jpg"))
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

```python
img.shape
```

```python
img_red = np.copy(img)
img_red[:, :, 1] = 0
img_red[:, :, 2] = 0
```

```python
plt.imshow(img_red)
plt.axis("off")
```

```python
fig, axes = plt.subplots(1, 1, figsize=(20, 20))
for sigma in range(4, 5, 1):
    img_blur = cv2.GaussianBlur(img_red, (5, 5), sigma)
    img_blur = cv2.GaussianBlur(img_blur, (5, 5), sigma)
    axes.imshow(img_blur)
    axes.axis("off")
```

```python
plt.imsave(fname="ssh_dark.png", arr=img_blur)
```
